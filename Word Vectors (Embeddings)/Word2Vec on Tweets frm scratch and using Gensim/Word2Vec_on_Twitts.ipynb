Creating more Data Inspection tools

```python
# Defining a function which calculates the total number of words that has a specific number of concurrent in the dataset.
# For e.g "concurrent(vocabs, 2) = 9847" means that there are 9847 words which have just 2 times of concurrent through the whole dataset.
def concurrent_number(dictionary, number_of_concurrents):
  try:
    number_of_concurrents[1]
    criteria = True
    x1 = number_of_concurrents[0]
    x2 = number_of_concurrents[1]
  except:
    criteria = False
  def concurrent_specific(dict, concurrents):
    sum = 0
    for w in dict.keys():
      if dict[w] == concurrents:
        sum += 1
    return sum

  if criteria == False:
    return concurrent_specific(dictionary, number_of_concurrents)
  if criteria == True:
    sum_occ = 0
    for i in range(x1,x2+1):
      sum_occ += int(concurrent_number(vocabs, i))

    return sum_occ
```

```python
# A function which enables you to know which words have a specific range of concurrents in the dataset.
# You can pass the "number of concurrents" both as a specific number or as an interval.
def concurrnet_words (dictionary, number_of_concurrents):
  try:
    number_of_concurrents[1]
    criteria = True
    x1 = number_of_concurrents[0]
    x2 = number_of_concurrents[1]
  except:
    criteria = False
  if criteria == True:
    for w in dictionary.keys():
      if (dictionary[w] > x1) and (dictionary[w] < x2):
        print(w)
  if criteria == False:
    for w in dictionary.keys():
      if dictionary[w] == number_of_concurrents:
        print(w)
```

Continuous Bag of Words (CBOW) moving window function:

```python
def cbow_xy_maker(dataset, window_size=7):
  import math
  half_window = math.floor(window_size/2)
  x, y= [], []
  for sentence in dataset:
    sentence = sentence.split()
    if len(sentence) >= window_size:
        for w in range(half_window, len(sentence)-half_window):
          features = sentence[w-half_window : w+half_window+1]
          target = features[half_window]
          del features[half_window]
          x.append(features)
          y.append(target)

  return x, y
```

```python
def get_embedding(embedding_matrice, word, word2tok):
  word_indice = word2tok[word]
  embedding_dim = embedding_matrice.shape[1]
  vocab_size = embedding_matrice.shape[0]
  word_onehot = np.zeros(shape=(vocab_size))
  word_onehot[word_indice] = 1.0
  wordvec = np.matmul(word_onehot, embedding_matrice)

  return wordvec
```

```python
def closest_words(word, embedding_matrice, word2tok, n=10):
  emb_target_word = get_embedding(embedding_matrice, word, word2tok)
  closest = []
  
  for counter, i in enumerate(word2tok.keys()):
    emb_word = get_embedding(embedding_matrice, i, word2tok)
    dist = np.linalg.norm(emb_target_word - emb_word)
    if len(closest) < 10:
      closest.append((i,float(f'{dist:.5f}')))
    else:
      if dist < max([i[1] for i in closest]):
        closest.append((i,float(f'{dist:.5f}')))
      closest = sorted(closest, key= lambda x: x[1])
      closest = closest[:10]
    if counter % 200 == 0:
      print(f'{counter}/{len(word2tok)}', closest, sep=' => ')

  return closest
```
